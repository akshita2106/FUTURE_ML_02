```python
import pandas as pd

df = pd.read_csv("../data/spotify_processed.csv")

```

```python
df.head()
```

```python
# Low music engagement
df['low_music_engagement'] = df['music_lis_frequency'].isin(
    ['Rarely', 'Occasionally']
).astype(int)

# Low podcast engagement
df['low_podcast_engagement'] = df['pod_lis_frequency'].isin(
    ['Never', 'Rarely']
).astype(int)

# Low recommendation satisfaction
df['low_recommendation_rating'] = (df['music_recc_rating'] <= 2).astype(int)

# No premium interest
df['no_premium_interest'] = (df['premium_sub_willingness'] == 'No').astype(int)

```

```python
df[
    [
        'low_music_engagement',
        'low_podcast_engagement',
        'low_recommendation_rating',
        'no_premium_interest'
    ]
].head()
```

```python
# Engagement score
df['engagement_score'] = (
    df['low_music_engagement'] +
    df['low_podcast_engagement']
)

# Satisfaction score
df['satisfaction_score'] = df['low_recommendation_rating']

# Monetization resistance score
df['monetization_score'] = df['no_premium_interest']

# Overall churn pressure
df['churn_pressure'] = (
    0.5 * df['engagement_score'] +
    0.3 * df['satisfaction_score'] +
    0.2 * df['monetization_score']
)
```

```python
df[
    [
        'engagement_score',
        'satisfaction_score',
        'monetization_score',
        'churn_pressure'
    ]
].head()
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.to_csv("../data/spotify_model_ready.csv", index=False)
```
